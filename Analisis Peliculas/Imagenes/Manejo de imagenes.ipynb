# Adicion de figuras a la imagen 

En la siguiente imagen creamos dos rectangulos para clsificar diferentes caballos

```python
imgHorse = np.array(Image.open('horses.jpg')) 

# Habilitamos los ejes en la imagen  
fig,ax = plt.subplots(1) 

# Visualizacion  
ax.imshow(imgHorse) 

# Creaccion del rectangulos
rect = patches.Rectangle((0,180),315,250,linewidth=3,edgecolor='r',facecolor='none',label='Caballo Blanco') 
rect2 = patches.Rectangle((420,200),185,150,linewidth=3,edgecolor='y',facecolor='none') 

# Add the patch to the Axes 
ax.add_patch(rect) 
ax.add_patch(rect2)

#Añadimos las etiquetas 
ax.text(30,170,'Caballo Blanco',fontsize=15)
ax.text(410,190,'Caballo Gris',fontsize=15)
```
